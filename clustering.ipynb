Q-1. What is unsupervised learning in the context of machine learning?

Ans-

- Unsupervised learning is a type of machine learning where the model is trained on data without labels.
- The algorithm tries to find patterns or structure in the data by itself.

Key Points:
- No target/output variable is provided
- Works only with input data
- Used for exploration and pattern discovery

Q-2. How does K-Means clustering algorithm work?

Ans-

K-Means is an unsupervised learning algorithm used to divide data into K clusters.

Steps:
- Choose number of clusters (K)
- Select K random points as centroids
- Assign each data point to the nearest centroid
- Calculate new centroids by taking the mean of each cluster
- Repeat steps 3 and 4 until centroids do not change

Key Concept:
- Minimizes distance between data points and their cluster center (centroid)

Q-3. Explain the concept of a dendrogram in hierarchical clustering.

Ans- A dendrogram is a tree-like diagram used to show how data points are grouped in hierarchical clustering.

Key Idea:

- It represents how clusters are merged (or split) step by step
- Shows the hierarchy of clusters

How It Works:
- Start with each data point as a separate cluster
- Merge the closest clusters step by step
- Continue until all points form one big cluster
- The dendrogram visually records these merges

Q-4.  What is the main difference between K-Means and Hierarchical Clustering?

Ans-

- K-Means is a partition-based algorithm that divides data into a fixed number of clusters (K). You must decide the value of K before running the algorithm. It works by repeatedly assigning data points to the nearest centroid and updating those centroids.
- Hierarchical Clustering is a tree-based algorithm that builds a hierarchy of clusters. It does not require specifying the number of clusters in advance. Instead, it creates a dendrogram, and you can choose the number of clusters later by cutting the tree at a certain level.

Q-5. What are the advantages of DBSCAN over K-Means?

Ans-
- No need to specify number of clusters (K)-
DBSCAN automatically finds the number of clusters, while K-Means requires you to choose K in advance.
- Can detect arbitrary-shaped clusters-
DBSCAN can find clusters of any shape (e.g., curved or irregular), whereas K-Means works best with spherical (round) clusters.
- Handles noise and outliers well-
DBSCAN can identify and label noise points (outliers), while K-Means forces every point into a cluster.
- More robust to initialization-
DBSCAN does not depend on random centroid initialization like K-Means, so results are more stable.
- Works well with dense regions-
DBSCAN groups points based on density, making it effective when clusters are defined by dense areas separated by sparse regions.

Q-6. When would you use Silhouette Score in clustering?

Ans-

The Silhouette Score is used to evaluate how well your clustering is performing.

When to Use It:
- When you want to check the quality of clusters
- When you need to compare different clustering algorithms
- When you want to find the best number of clusters (K) in algorithms like K-Means
- When you want to ensure clusters are well-separated and compact

Q-7. What are the limitations of Hierarchical Clustering?

Ans-

- Computationally expensive-
It has high time and memory complexity, so it becomes very slow for large datasets.
- Not scalable-
Works well for small datasets but is not suitable for big data.
- Once merged/split, cannot be undone-
In hierarchical clustering, decisions are final. If a wrong merge happens early, it cannot be corrected later.
- Sensitive to noise and outliers-
Outliers can significantly affect the clustering structure.
- Choice of distance metric matters-
Different distance measures (Euclidean, Manhattan, etc.) can lead to very different results.
- Difficult to interpret for large datasets-
The dendrogram becomes complex and hard to analyze when the number of data points increases.

Q-8. Why is feature scaling important in clustering algorithms like K-Means?

Ans-

Feature scaling is important because K-Means uses distance (usually Euclidean distance) to form clusters.

Why It Matters:
- Prevents dominance of large-scale features-
Features with bigger values (e.g., income in lakhs) can dominate smaller ones (e.g., age), leading to biased clustering.
- Ensures fair contribution of all features-
Scaling makes sure each feature contributes equally to distance calculation.
- Improves clustering accuracy-
Without scaling, clusters may be formed incorrectly due to uneven feature influence.
- Faster convergence-
Properly scaled data helps K-Means converge (stabilize) more quickly.

Q-9. How does DBSCAN identify noise points?

Ans-

DBSCAN (Density-Based Spatial Clustering) classifies points into:
- Core points
- Border points
- Noise points (outliers)

Key Parameters:
- ε (epsilon) → Radius (neighborhood distance)
- MinPts → Minimum number of points required to form a dense region

Q-10. Define inertia in the context of K-Means.

Ans-

Inertia is the sum of squared distances between each data point and its nearest cluster centroid

Key Points:
- Measures how close data points are to their cluster center
- Used to evaluate clustering quality
- Lower inertia → better, compact clusters
- Higher inertia → poor clustering

Q-12. Describe the concept of "density" in DBSCAN.

Ans- Density refers to how closely packed data points are in a region.

Key Idea:
- A region is considered dense if it has many points within a given radius (ε)
- A region is sparse if it has few points

Parameters Used:
- ε (epsilon) → Radius around a point
- MinPts → Minimum number of points required to form a dense region

Q-13. Can hierarchical clustering be used on categorical data?

Ans- Yes, hierarchical clustering can be used on categorical data, but not directly with standard distance measures like Euclidean distance.

Key Points:
- Categorical data needs special distance/similarity measures
- Common methods:
- Hamming Distance
- Jaccard Similarity

Q-14. What does a negative Silhouette Score indicate?

Ans- A negative Silhouette Score indicates poor clustering.

Key Points:
- Data point is closer to another cluster than its own cluster
- Suggests wrong cluster assignment
- Clusters are overlapping or not well separated

Q-15. Explain the term "linkage criteria" in hierarchical clustering.

Ans- Linkage criteria defines how the distance between two clusters is calculated in hierarchical clustering.

Key Idea:
- When merging clusters, we need a rule to decide which clusters are closest
Linkage criteria provides that rule
Common Types:
- Single Linkage-
Distance between closest points of two clusters
- Complete Linkage-
Distance between farthest points of two clusters
- Average Linkage-
Average distance between all points of two clusters
- Centroid Linkage-
Distance between centroids (means) of clusters

Q-16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans- K-Means assumes clusters are similar in size and density

Key Problems:
1.Assumes equal-sized clusters-
Struggles when one cluster is large and another is small

2.Assumes uniform density-
Cannot handle clusters where some are dense and others are sparse

3.Uses distance to centroid-
Points may be assigned incorrectly if clusters overlap or vary in spread

4.Sensitive to outliers-Outliers can shift centroids, affecting cluster formation

5.Fails on non-spherical shapes-
Works best for round clusters, not irregular ones

Q-17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans- DBSCAN mainly uses two parameters to form clusters:

1.ε (Epsilon)
- Radius to find nearby points
- Small ε → more noise
- Large ε → bigger/merged clusters

2.MinPts
- Minimum points to form a cluster
- Small MinPts → more clusters
- Large MinPts → more noise

Q-18. How does K-Means++ improve upon standard K-Means initialization?

Ans- K-Means++ improves how initial centroids are chosen

Key Points:
- Standard K-Means uses random initialization
- K-Means++ selects centroids more carefully (far apart from each other)

How It Helps:
- Reduces chances of poor clustering
- Avoids bad initial centroid positions
- Leads to better and more stable results
- Often faster convergence

Q-19. What is agglomerative clustering?

Ans- Agglomerative clustering is a type of hierarchical clustering

Key Idea:
- It follows a bottom-up approach
- Starts with each data point as a separate cluster

Steps:
- Treat each point as its own cluster
- Find the closest two clusters
- Merge them into one cluster
- Repeat until all points form a single cluster

Q-20.  What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans- Silhouette Score is better than inertia because it considers both cluster compactness and separation

Key Points:
1.Inertia only measures:
- How close points are to their own cluster centroid
- Lower value = better

2.Silhouette Score measures:
- How close a point is to its own cluster
- Compared to other clusters

Why Silhouette is Better:
- Considers both intra-cluster and inter-cluster distance
- Detects overlapping clusters
- Provides a clear range (-1 to +1) for evaluation
- Helps better in choosing optimal number of clusters

```python
# Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

k_values = range(2, 6)
sil_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")


plt.figure(figsize=(8, 5))
plt.bar(k_values, sil_scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K-Means (k = 2 to 5)")
plt.xticks(k_values)
plt.show()
```

```python
# Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=2.5,
    random_state=42
)


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)


h = 0.1
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)


Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3)


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, edgecolor='k')


centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


data = load_breast_cancer()
X = data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)
    print(f"Inertia for K={k}: {kmeans.inertia_:.2f}")


plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia_values, marker='o')
plt.title("Inertia Values for K-Means (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
#  Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D


digits = load_digits()
X = digits.data
y = digits.target


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)


kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_pca)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],
    c=labels, cmap='tab10', s=40
)


centers = kmeans.cluster_centers_
ax.scatter(
    centers[:, 0], centers[:, 1], centers[:, 2],
    c='red', s=250, marker='X', label='Centroids'
)

ax.set_title("KMeans Clustering on Digits Dataset (PCA to 3D)")
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.legend()

plt.show()
```

```python
# Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the clustering

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, y_true = make_blobs(n_samples=400, centers=5, cluster_std=1.0, random_state=42)


kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)


score = silhouette_score(X, labels)

print("Silhouette Score:", score)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=250, marker='X', label='Centroids')

plt.title("KMeans Clustering with 5 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering


data = load_breast_cancer()
X = data.data
y = data.target


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agg.fit_predict(X_pca)


plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50, edgecolor='k')
plt.title("Agglomerative Clustering on Breast Cancer Dataset (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
#  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)


sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)
avg_silhouette = silhouette_score(X_scaled, cluster_labels)

print("Average Silhouette Score:", avg_silhouette)


plt.figure(figsize=(10, 6))
y_lower = 10

for i in range(n_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values
    )

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10


plt.axvline(x=avg_silhouette, linestyle='--', label='Average Silhouette Score')

plt.title("Silhouette Plot for Iris Dataset using KMeans")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.yticks([])
plt.legend()
plt.show()
```

```python
# Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.2, random_state=42)

# Step 2: Apply Agglomerative Clustering with average linkage
agg = AgglomerativeClustering(n_clusters=3, linkage='average')
labels = agg.fit_predict(X)

# Step 3: Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50, edgecolor='k')
plt.title("Agglomerative Clustering with Average Linkage")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

# Load the Wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Create a DataFrame
df = pd.DataFrame(X, columns=feature_names)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Wine dataset has 3 classes
df['Cluster'] = kmeans.fit_predict(X)

# Visualize the first 4 features with cluster assignments
sns.pairplot(df.iloc[:, :4].join(df['Cluster']), hue='Cluster', palette='Set2')
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
#  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count
# Import libraries
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Add some random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(20, 2))  # 20 random noise points
X_noisy = np.vstack([X, noise])

# Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_noisy)

# Count clusters and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

```python
# Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import seaborn as sns

# Load Digits dataset
digits = load_digits()
X = digits.data
y_true = digits.target

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, learning_rate='auto')
X_2d = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10)  # Digits dataset has 10 classes
labels = agg.fit_predict(X_2d)

# Create a DataFrame for plotting
df = pd.DataFrame(X_2d, columns=['TSNE1', 'TSNE2'])
df['Cluster'] = labels

# Plot clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='TSNE1', y='TSNE2', hue='Cluster', palette='tab10', s=60)
plt.title('Agglomerative Clustering on Digits Dataset (t-SNE Reduced)')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
